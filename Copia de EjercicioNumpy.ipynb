```python
import numpy as np
```

```python
edades = np.random.uniform(
    low= 20,
    high= 50,
    size= (3,3)
)
matriz_edades = np.round(edades)
matriz_edades
```

```python
ingresos = np.random.uniform(
    low= 2000,
    high= 5000,
    size= (3,3)
)
matriz_ingresos = np.round(ingresos)
matriz_ingresos
```

```python
array_edades = matriz_edades.reshape(9,1)
array_ingresos = matriz_ingresos.reshape(9,1)
edades_ingresos = np.column_stack((array_edades, array_ingresos))
display(
    array_edades,
    array_ingresos,
    edades_ingresos
)
```

```python
bool_edades = np.logical_and(array_edades > 20, array_edades < 30)
bool_inger = np.logical_and(array_ingresos > 2000, array_ingresos < 5000)
bool_matr = np.column_stack((bool_edades, bool_inger))
display (
    bool_edades,
    bool_inger,
    bool_matr
)
```

```python
validaCumple = np.logical_and(bool_edades, bool_inger)
validaCumpleColum = np.column_stack((validaCumple, validaCumple))
validaCumpleColum
```

```python
print("Los registros que cumple las condiciones son: ")
edades_ingresos[validaCumpleColum]
```
