```python
indicadores = pd.read_csv('/content/Colab/indicadores-mundiales.csv')
```

```python
indicadores.shape
indicadores.columns
```

```python
promVidaH = indicadores.groupby(['Region'])['Esperanza de vida (varones)'].mean()
promVidaH
```

```python
promVidaM = indicadores.groupby(['Region'])['Esperanza de vida (mujeres)'].mean()
promVidaM
```

```python
IndPromVidaH = indicadores.groupby(['Region'])['Esperanza de vida (varones)'].transform('mean')
indicadores['PromVidaH'] = IndPromVidaH
indicadores
```

```python
IndpromVidaM = indicadores.groupby(['Region'])['Esperanza de vida (mujeres)'].transform('mean')
indicadores['promVidaM'] = IndpromVidaM
indicadores
```

```python
Africa = indicadores[indicadores['Region'] == 'africa']
Africa
```

```python
esp_vida_menor_reg = Africa[Africa["Esperanza de vida (mujeres)"] < Africa["promVidaM"]]
esp_vida_menor_reg
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(esp_vida_menor_reg['Esperanza de vida (mujeres)'], esp_vida_menor_reg['Pais'], color='b')
plt.plot()

plt.xlabel("Esperanza de vida (mujeres)")
plt.ylabel("Pais")
plt.title("Esperanza de vida (mujeres) menor al promedio Africa")
plt.show()
```
